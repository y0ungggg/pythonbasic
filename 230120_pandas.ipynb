```python
# fillna() 메서드를 사용하여 NaN 값을 원하는 값으로 바꿀 수 있습니다. 첫 인자로 NaN을 변경하고자 하는 값을 전달하면 됩니다.
df = pd.DataFrame([[np.nan,2,np.nan,0],
                  [3,4,np.nan,1],
                  [np.nan,np.nan,np.nan,np.nan],
                  [np.nan,3,np.nan,4]],
                 columns=list('ABCD'))
df
df.fillna(0)
```

```python
values = {'A' : 0, 'B':1, 'C':2, 'D':3}
df.fillna(value=values)
```

```python
# fillna() 메서드에 limit 키워드 인자에 숫자를 전달하여 그 숫자만큼 column마다 변경 횟수를 제한할 수 있습니다.
values = {'A' : 0, 'B':1, 'C':2, 'D':3}
df.fillna(value=values, limit = 1)
```

```python
# fillna() 메서드에 limit 키워드 인자에 숫자를 전달하여 그 숫자만큼 column마다 변경 횟수를 제한할 수 있습니다.
df2 = pd.DataFrame(np.zeros((3, 4)), columns=list('ABCE'))
df.fillna(df2)
df
```

```python
ages = [0,2,10,21,23,37,31,61,20,41,32,101]
```

```python
bins = [1, 20, 30, 50, 70, 100] # 구간
labels = ['미성년자', '청년', '장년', '중년', '노년'] # 라벨 
cats = pd.cut(ages, bins, labels=labels)
cats
```

```python
cats.codes
```

```python
df4 = pd.DataFrame(ages, columns=['ages'])
df4['age_cat'] = pd.cut(df4.ages, bins, labels=labels)
df4
```
